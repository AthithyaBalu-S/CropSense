```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from pycaret.classification import *
```

```python
model = setup(data = data, target = 'Crop' , categorical_features = ['Nitrogen','Phosphorus','Potassium','Temperature','Humidity','pH_Value','Rainfall'])
```

```python
best_model = compare_models()
```

```python
x=data.iloc[:,:-2]
y=data.iloc[:,-2]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
print(confusion_matrix(y_test,y_pred))
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=100)
rfc.fit(x_train,y_train)
```

```python
y_pred1 = rfc.predict(x_test)
```

```python
print(confusion_matrix(y_test,y_pred1))
```

```python
print(accuracy_score(y_test,y_pred1))
```

```python
print(classification_report(y_test,y_pred1))
```

```python
from sklearn.linear_model import RidgeClassifier
```

```python
ridge_clf = RidgeClassifier()
```

```python
ridge_clf.fit(x_train, y_train)
```

```python
y_pred_ridge = ridge_clf.predict(x_test)
```

```python
print(confusion_matrix(y_test, y_pred_ridge))
```

```python
print(accuracy_score(y_test, y_pred_ridge))
```

```python
print(classification_report(y_test, y_pred_ridge))
```

```python
from sklearn.ensemble import VotingClassifier
```

```python
ensemble_model = VotingClassifier(estimators=[('lr', model), ('rf', rfc), ('rc', ridge_clf)], voting='hard')
```

```python
ensemble_model.fit(x_train, y_train)
```

```python
y_pred_ensemble = ensemble_model.predict(x_test)
```

```python
print(confusion_matrix(y_test, y_pred_ensemble))
```

```python
print(accuracy_score(y_test, y_pred_ensemble))
```

```python
print(classification_report(y_test, y_pred_ensemble))
```
